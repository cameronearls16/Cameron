```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("../../course_assignments/homework7/star_data.txt", skiprows=1)
print("Shape:", data.shape)  # (rows, 2)

magnitude   = data[:, 0]
temperature = data[:, 1]
```

```python

idx_brightest = np.argmin(magnitude)
idx_dimmest   = np.argmax(magnitude)
idx_hottest   = np.argmax(temperature)
idx_coldest   = np.argmin(temperature)

print("Magnitude min/max (brightest/dimmest):", magnitude.min(), magnitude.max())
print("Temperature min/max (coldest/hottest):", temperature.min(), temperature.max())
print("Indices -> brightest, dimmest, hottest, coldest:",
      idx_brightest, idx_dimmest, idx_hottest, idx_coldest)
```

```python
log_magnitude   = np.log(magnitude)
log_temperature = np.log(temperature)

plt.figure(figsize=(8, 6))
plt.scatter(log_temperature, log_magnitude, s=15, alpha=0.7, label="Stars")

plt.scatter(log_temperature[idx_brightest], log_magnitude[idx_brightest],
            color="black", marker="*", s=300, label="Brightest")
plt.scatter(log_temperature[idx_dimmest],   log_magnitude[idx_dimmest],
            color="purple", marker="*", s=300, label="Dimmest")
plt.scatter(log_temperature[idx_hottest],   log_magnitude[idx_hottest],
            color="green",  marker="*", s=300, label="Hottest")
plt.scatter(log_temperature[idx_coldest],   log_magnitude[idx_coldest],
            color="orange", marker="*", s=300, label="Coldest")

ax = plt.gca()
ax.invert_xaxis()
ax.invert_yaxis()

plt.title("Hertzsprung-Russel Diagram")
plt.xlabel("Temperature (log K)")
plt.ylabel("Magnitude")
plt.legend()
plt.tight_layout()
plt.show()
```
