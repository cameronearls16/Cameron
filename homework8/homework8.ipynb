```python
median_lat = df["latitude"].median()
median_lon = df["longitude"].median()
```

```python
df['lat_distance'] = abs(df['latitude'] - median_lat)
df['lon_distance'] = abs(df['longitude'] - median_lon)
df.shape
```

```python
closest_lat_idx = df["lat_distance"].idxmin()
closest_lat_state = df.loc[closest_lat_idx, "state"]
closest_lat_dist = df.loc[closest_lat_idx, "lat_distance"]

print(closest_lat_idx, closest_lat_state, closest_lat_dist)
```

```python
farthest_lat_idx = df["lat_distance"].idxmax()
farthest_lat_state = df.loc[farthest_lat_idx, "state"]
farthest_lat_dist = df.loc[farthest_lat_idx, "lat_distance"]

print(f"The state farthest from the median latitude is {farthest_lat_state} at {farthest_lat_dist:.2f} degrees away.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("dowjones")
```

```python
df = df.dropna(axis=1)
```

```python
print("Shape:", df.shape)
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

print("\nFirst five rows:\n", df.head())
print("\nLast five rows:\n", df.tail())
print("\nColumn names:", list(df.columns))
```

```python
#Date is the trading date and Price is the closing price
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

df["Date_num"] = range(len(df))

sns.scatterplot(data=df, x="Date_num", y="Price", ax=axes[0, 0])
axes[0, 0].set_title("Dow Jones Price Over Time")
axes[0, 0].set_xlabel("Time (index of dates)")
axes[0, 0].set_ylabel("Price")

sns.lineplot(data=df, x="Date_num", y="Price", ax=axes[0, 1], color="orange")
axes[0, 1].set_title("Trend of Dow Jones Price Over Time")

sns.histplot(data=df, x="Price", kde=True, ax=axes[1, 0], color="skyblue")
axes[1, 0].set_title("Distribution of Dow Jones Prices")

sns.boxplot(data=df, y="Price", ax=axes[1, 1], color="lightgreen")
axes[1, 1].set_title("Boxplot of Dow Jones Prices")

plt.tight_layout()
plt.show()
```
